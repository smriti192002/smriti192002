#Missing value analysis for each columns

#cigsperday

```python
heart_data[heart_data['cigsPerDay'].isnull()]
```

```python
import math
heart_data[(heart_data['is_smoking']=='YES') & (heart_data['cigsPerDay'].notnull()) &
(heart_data['cigsPerDay']<60)].groupby('TenYearCHD').aggregate({'cigsPerDay':'mean'})
```

```python
impute_cigperday_0=math.floor(heart_data[(heart_data['is_smoking']=='YES') & (heart_data['cigsPerDay'].notnull()) &
(heart_data['cigsPerDay']<60)].groupby('TenYearCHD').aggregate({'cigsPerDay':'mean'}).iloc[0])

impute_cigperday_1=math.floor(heart_data[(heart_data['is_smoking']=='YES') & (heart_data['cigsPerDay'].notnull()) &
(heart_data['cigsPerDay']<60)].groupby('TenYearCHD').aggregate({'cigsPerDay':'mean'}).iloc[1])
```

```python
heart_data['cigsPerDay']=heart_data.apply(lambda x: impute_cigperday_0 if (x['TenYearCHD']==0 and np.isnan(x['cigsPerDay']))
else impute_cigperday_1 if (x['TenYearCHD']==1 and np.isnan(x['cigsPerDay'])) else x['cigsPerDay'], axis=1 )
```
